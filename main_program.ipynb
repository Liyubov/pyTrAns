# Modules for data analysis

For continuous trajectories: 
1. Calculation of scaling exponent for trajectories
2. Calculation of Hurst exponent 
3. Calculation of MSD: the moments of displacements of each trajectory can provide unique identifies for trajectories. For example, connected with the diffusion, we are mainly interested with the second moment of the distribution of trajectory's displacement, called, mean squared displacement (MSD).


For discrete trajectories on networks: 
1. Estimation of Markov matrix (for trajectory on a network) 
2. Calculation of propagator 
3. Calculation of stationary distribution and other spectral properties of Markov matrix
